```python
#Covariance: degree_p to etest_p
#Value: 21.929469
#Interpretation: This is a positive covariance. It suggests that, in general, as a student's degree percentage (degree_p) increases, their employment test percentage (etest_p) also tends to increase. However, the strength of this relationship needs to be considered in the context of the variables' scales.
#Covariance: etest_p to mba_p
#Value: 16.886973
#Interpretation: Again, this is a positive covariance. It indicates that there's a tendency for higher employment test percentages (etest_p) to be associated with higher MBA percentages (mba_p). Like before, the magnitude of the covariance should be interpreted with caution.
```
